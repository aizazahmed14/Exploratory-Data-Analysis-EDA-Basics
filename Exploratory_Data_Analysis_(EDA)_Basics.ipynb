```python
#Understanding Distributions, Relationships and tredns

import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

data = '/Users/aizaz/Downloads/Salary_Data.csv'

df = pd.read_csv(data)
dataf = pd.DataFrame(df)
print(dataf.head(10))
```

```python
#Basic stats
print(dataf.describe())
```

```python
#Visualising the salary Distribution

plt.figure(figsize=(10,6))
sns.histplot(df['Salary'], kde=True, bins=10, color='blue')
plt.title('Salary Distribution')

plt.xlabel('Salary')

plt.ylabel('Frequency')

plt.show()
```

```python
#Scatterplot for relationships of the years of experience

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Years of Experience", y='Salary', hue='Gender', style='Education Level', size='Age', palette='cool')
plt.title('Years of Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
plt.show()
```

```python
#interactive bar chart for job titles and salaries 

fig = px.bar(df, x='Job Title', y='Salary', color = 'Education Level',title='Average Salary by job Title')
fig.show()
```
